## Selenium e WebDriver
Como atualmente o site do GeoSampa possui muitos links e processos JavaScript vamos optar por usar o Selenium para cessar diretamente através de uma instância de um browser, um WebDriver.

```python
from selenium import webdriver

options = webdriver.ChromeOptions()

# Descomente a linha abaixo caso não precise visualizar o browser
# options.add_argument('headless') 

options.add_argument('window-size=900x1080')

# Inicializando o driver, que é uma instância de um browser
driver = webdriver.Chrome(options=options)
driver.get("http://geosampa.prefeitura.sp.gov.br/PaginasPublicas/_SBC.aspx")
```

```python
layers =  driver.find_element_by_xpath('//*[@id="LayersWhiteCamadas"]')
layers
```

```python
layers_for_expand = driver.find_elements_by_class_name('lblLayerSwitcherItem')

#for l in layers_for_expand:
#    print(l.text)
```

```python
btns_mais_menos = driver.find_elements_by_xpath('//*[contains(@id, "btnMaisMenos")]')
for b in btns_mais_menos:
    b.click()
```

```python
from pprint import pprint

layers = driver.find_elements_by_xpath('//*[contains(@class, "lblLayerSwitcherItem")]')
for l in layers:
    #print(l.id)
    i = l.find_element_by_xpath('.//preceding-sibling::input')
    #print(i.get_attribute('id'))
    print(i.get_attribute('indice'))
    print(l.get_attribute('innerHTML'))
```

## Gravando os dados

```python
f = open("inventario/inventario_camadas_geosampa_site.txt","w+")
f.write("Seguem a seguir as %s camadas, grupos e subgrupos obtidas por WebScraping do site do GeoSampa\n" %(str(len(layers))))

grupos = 0
subgrupos = 0
camadas = 0

for l in layers:
    f.write("*" * 30)
    f.write("\n")
    indice_ou_tipo = l.find_element_by_xpath('.//preceding-sibling::input').get_attribute('indice')
    f.write("Índice: %s\n" %(indice_ou_tipo))
    if indice_ou_tipo == 'G':
        grupos+=1
        f.write("Grupo: %s\n" %(l.get_attribute('innerHTML')))
    elif indice_ou_tipo == 'S':
        subgrupos+=1
        f.write("SubGrupo: %s\n" %(l.get_attribute('innerHTML')))
    else:
        camadas+=1
        f.write("Camada: %s\n" %(l.get_attribute('innerHTML')))

f.write("*" * 30)
f.write("\n")
f.write("Total de grupos:    %s\n" %(str(grupos)))
f.write("Total de subgrupos: %s\n" %(str(subgrupos)))
f.write("Total de camadas:   %s\n" %(str(camadas)))
        
f.close()
```
